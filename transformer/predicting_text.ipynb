```python
import sys
sys.path.append("/home/surchand/projects/TranslateManipuri/src")
from excel_to_text import process_string
#translator_model_name = "translate_manipuri_model/3"
translator_model_name = "translate_manipuri_model/gpu-4l-20epoch"
```

```python
reloaded = tf.saved_model.load(translator_model_name)
```

```python
d = reloaded(["The barn was made of wood."])
```

```python
print(d["prediction"][0].numpy().decode('utf-8'))
```

```python
readb = open(file_location, "r")
writeb = open(out_file, "w+")
for i in readb:
    output = reloaded([i])
    decoded_output = output["prediction"][0].numpy().decode('utf-8')
    result = process_string(decoded_output)
    print("i = {}  o = {}".format(i, result))
    writeb.write(result + "\n")
```
