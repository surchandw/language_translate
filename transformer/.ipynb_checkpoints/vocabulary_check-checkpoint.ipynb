```python
def tokenize_words(text):
    # Remove punctuation
    # Add Meitei mayek ꯫
    remove_chars = string.punctuation + "꯫"
    text = text.translate(str.maketrans("", "", remove_chars))
    
    # Split the text into words
    words = text.split()

    return words
```

```python
#example = manipuri_se
tokenized_sents = [tokenize_words(i) for i in manipuri_se]
word_dict = dict()
for list_item in tokenized_sents:
    for word in list_item:
        if word not in word_dict.keys():
            word_dict[word] = 1
        else:
            word_dict[word] += 1
sorted_dict = dict(sorted(word_dict.items()))
with open("vocab_verification", "w") as f:
    for key in sorted_dict:
        f.write("{} => {}\n".format(key, sorted_dict[key]))
    f.close()   
```
